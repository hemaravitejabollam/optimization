**QUESTION #1**

```python
# Import pyomo package
from pyomo.environ import *

# Create a model
model= ConcreteModel()

# declare variables
model.x1 = Var(domain=NonNegativeReals)
model.x2 = Var(bounds =(0,10))

# objective function
model.profit = Objective(expr = 2*model.x1 + model.x2, sense=maximize)

#constraints
model.constraint1 = Constraint(expr = 2*model.x1 + 5*model.x2 <= 60)
model.constraint2 = Constraint(expr = model.x1 + model.x2 <=18)
model.constraint3 = Constraint(expr = 3*model.x1 + model.x2 <=44)

#solve
results = SolverFactory('glpk').solve(model)
results.write()
model.pprint()

# display solutions
print('\nProfit = ', model.profit())

print('\nx1 = ' , model.x1())
print('x2 = ' , model.x2())
```

**QUESTION #2: SilverCo Company**

```python
# Import pyomo package
from pyomo.environ import *

# Create a model
model= ConcreteModel()

# declare variables
model.x1 = Var(domain=NonNegativeReals)
model.x2 = Var(domain=NonNegativeReals)
model.x3 = Var(domain=NonNegativeReals)
model.x4 = Var(domain=NonNegativeReals)
model.x5 = Var(domain=NonNegativeReals)

# objective function
model.cost = Objective(expr = 22*model.x1 + 20*model.x2 + 25*model.x3 + 24*model.x4 + 27*model.x5, sense=minimize)

#constraints
model.constraint1 = Constraint(expr = 60*model.x1 + 25*model.x2 + 45*model.x3 + 20*model.x4 + 50*model.x5 == 40)
model.constraint2 = Constraint(expr = 10*model.x1 + 15*model.x2 + 45*model.x3 + 50*model.x4 + 40*model.x5 == 35)
model.constraint3 = Constraint(expr = 30*model.x1 + 60*model.x2 + 10*model.x3 + 30*model.x4 + 10*model.x5 == 25)

#solve
results = SolverFactory('glpk').solve(model)
results.write()
model.pprint()

# display solutions
print('\nCost = ', model.cost())

print('\nx1 = ' , model.x1())
print('x2 = ' , model.x2())
print('x3 = ' , model.x3())
print('x4 = ' , model.x4())
print('x5 = ' , model.x5())
```

**QUESTION #4: Logan Manufacturing**

```python
# Import pyomo package
from pyomo.environ import *

# Create a model
model= ConcreteModel()

# declare variables
model.A = Var(bounds =(0,2200))
model.B = Var(bounds =(0,3600))


# objective function
model.cost = Objective(expr = 1.2*model.A + 1.7*model.B, sense=minimize)

#constraints
model.constraint1 = Constraint(expr = model.A + model.B >= 3300)
model.constraint2 = Constraint(expr = model.A + model.B <= 4300)
model.constraint3 = Constraint(expr = (88*model.A + 66*model.B) >= 78*(model.A+model.B))

#solve
results = SolverFactory('glpk').solve(model)
results.write()
model.pprint()

# display solutions
print('\nCost = ', model.cost())

print('\nA = ' , model.A())
print('B = ' , model.B())
```
